Anteriormente observamos estimados de nuestras variables númericas, sin embargo, una imágen vale más que mil palabras, en este postwork crearemos visualizaciones de nuestras distribuciones como boxplots, histogramas y gráficas de densidad. Primero importemos nuestros bibliotecas:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_clean = pd.read_csv("https://raw.githubusercontent.com/JMCastle98/M4_Analisis_de_Datos_con_Python/main/Dataset/pok%C3%A9dex_clean.csv", index_col=0)
```

Para facilitar la obtención de gráficos boxplot, se creó una pequeña función:

```python
def b0xpl0t(campo):
    sns.set(style="whitegrid")
    sns.boxplot(x=campo)
    plt.axvline(campo.mean(), c='y')

def estimados(campo):
    from scipy import stats
    print(f'El promedio de {campo.name} es: '+'{0:.3f}'.format(campo.mean()))
    print(f'La mediana de {campo.name} es: '+'{0:.3f}'.format(campo.median()))
    print(f'La media truncada de {campo.name} es: '+'{0:.3f}'.format(stats.trim_mean(campo, 0.1)))
    print(f'La desviación estándar de {campo.name} es: '+'{0:.3f}'.format(campo.std()))
    print(f'El máximo de {campo.name} es: '+'{0:.3f}'.format(campo.max()))
    print(f'El mínimo de {campo.name} es: '+'{0:.3f}'.format(campo.min()))
    print(f'El rango de {campo.name} es: '+'{0:.3f}'.format(campo.max() - campo.min()))
    print(f'El percentil 25 de {campo.name} es: '+'{0:.3f}'.format(campo.quantile(0.25))+' y el percentil 75 es: '+'{0:.3f}'.format(campo.quantile(0.75)))
    print(f'El rango intercuartílico de {campo.name} es: '+'{0:.3f}'.format(campo.quantile(0.75)-campo.quantile(0.25)))   
```

Combinada con la función anterior tenemos acceso a la información númerica y una visualización de estos:

```python
b0xpl0t(df_clean['height_m'])
estimados(df_clean['height_m'])
```

## IQR-Score (Interquartile Range Score)

Utilizamos el **Score de Rango Intercuartílico** como un filtro de los datos que exceden el tamaño de los bigotes de nuestro boxplot, de esta forma filtramos los valores atípicos y observamos mejor la concentración de nuestros datos. Para facilitar la tarea de obtener el IQR-Score, creamos dos funciones, la primera nos muestra los estimados númericos y el gráfico boxplot del dataset filtrado mediante IQR-Score. La segunda función por otra perte realiza el procedimiento de filtrado y nos devuelve el dataframe sin valores atípicos:


```python
def IQR(campo):
    iqr = df_clean[campo].quantile(0.75) - df_clean[campo].quantile(0.25)
    quantil25= df_clean[campo].quantile(0.25) - (iqr * 1.5)
    quantil75= df_clean[campo].quantile(0.75) + (iqr * 1.5)
    filtro_inferior = df_clean[campo] > quantil25
    filtro_superior = df_clean[campo] < quantil75
    
    df_filtrado = df_clean[filtro_inferior & filtro_superior]

    sns.set(style="whitegrid")
    sns.boxplot(x=df_filtrado[campo])
    plt.axvline(df_filtrado[campo].mean(), c='y')
    print(f'El IQR sin filtrar es: {iqr}')  
    print(f'El filtro inferior es: {quantil25} y el filtro superior es {quantil75}')
    estimados(df_filtrado[campo])

def IQR_df(campo):
    iqr = df_clean[campo].quantile(0.75) - df_clean[campo].quantile(0.25)
    quantil25= df_clean[campo].quantile(0.25) - (iqr * 1.5)
    quantil75= df_clean[campo].quantile(0.75) + (iqr * 1.5)
    filtro_inferior = df_clean[campo] > quantil25
    filtro_superior = df_clean[campo] < quantil75
    
    df_filtrado = df_clean[filtro_inferior & filtro_superior]

    return df_filtrado
```

## Tablas de frecuencia e histogramas

Son herramientas de análisis, una tabla de frecuencias divide nuestro conjunto de datos en segmentos de igual tamaño y muestra la cantidad de datos que existe en cada segmento. Un histograma es un gráfico de barras que representa una tabla de frecuencias, finalmente un gráfico de densidad es un versión suavizada de un histograma que nos muestra la densidad de probabilidad de nuestros datos.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_clean = pd.read_csv("https://raw.githubusercontent.com/JMCastle98/M4_Analisis_de_Datos_con_Python/main/Dataset/pok%C3%A9dex_clean.csv", index_col=0)
```

Vamos a obtener estos 3 resultados con funciones para facilitar el proceso:

```python
def frecuencia(campo,bin,nombre,color):
    #Tabla de frecuencias
    print(f'El rango es {campo.max() - campo.min()}')
    segmentos = pd.cut(campo, bin)
    print('La tabla de frecuencias es: ')
    print(campo.groupby(segmentos).count())
    
    #Utilizamos este paquete para ignorar el warning que causa la depreciación de distplot
    import warnings
    warnings.simplefilter(action='ignore', category=FutureWarning)

    #histograma y densidad
    fig, axes = plt.subplots(1, 2,figsize=(15, 5), sharey=False);
    fig.suptitle(f'Histograma y gráfico de densidad del campo: {nombre}', x= 0.3);

    from scipy.stats import skew, kurtosis 

    sns.set(style='whitegrid');
    sns.distplot(ax=axes[0],a=campo, kde=False, bins=bin, color=color);
    axes[0].set(ylabel= "conteo")
    axes[0].set(xlabel= nombre)
    plt.axvline(campo.mean(), c=color)

    sns.distplot(ax=axes[1],a=campo, hist=False, color = color);
    axes[1].set(ylabel= "densidad")
    axes[1].set(xlabel= nombre)
    axes[1].set_title('Curtosis:'+'{0:.3f}'.format(kurtosis(campo)) + '\n Asimetría:'+'{0:.3f}'.format(skew(campo))) #asimetria
    
```

Veamos los resultados de algunos campos númericos:

```python
data = IQR_df('height_m') #Utilizamos los datos filtrados mediante el IQR-Score
frecuencia(data['height_m'],10,'altura en metros', 'g')
```

Si siguieramos observando cada estadística de los pokémon, sería normal encontrarnos con una asímetria positiva, debido a que existen más pokémon normales o en etapas evolutivas medias que pokémono legendarios o en su forma final:

```python
fig = plt.figure()
fig.suptitle(f'Gráficos de densidad de estadísticas pokémon');
sns.set(style='white')

for campo in df_clean[['sp_attack','sp_defense','defense','speed']].columns:
    sns.distplot(IQR_df(campo)[campo], hist = False, kde_kws = {'shade': True});

plt.ylabel("Densidad")
plt.xlabel("Puntos")    
```
